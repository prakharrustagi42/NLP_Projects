```python
## our final aim moving forward is to create a bag of words for each message 
## words preprocessing 

import string
s_message = "Sample message :Notice it has punctuations!."
```

```python
## making a function for the entire thing :

def word_processing(message):
    '''
        Remove the punctuations
        remove the stop words
        return clean list of words
    '''
    
    no_punc = [char for char in message if char not in string.punctuation]
    no_punc = "".join(no_punc)
    cleaned_message = [word for word in no_punc.split() if word.lower not in stopwords.words("english")]
    return cleaned_message
```

```python
word_processing("Sample message :Notice it has punctuations!.")

## this function here is manual tokenization .. i.ee to get words out of sentences in a list 
```

```python
## trying this with first 5 text messages 


messages["message"].head().apply(lambda x : word_processing(x))

## here the steps covered are tokenization and then we usually normalize that is converting to base form 
## example ran run, running all become run and that is called stemming 
## but here we have a lot of short words like U, c etc which dont work well with stemming 
```

```python
## Vectorization : this is where we convert this list in to a bag of words model so that ML can understand this 

from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer = word_processing)

bow_transformer.fit(messages["message"])
```

```python
## so the whole chage is to 

# 1. Remove the entire punctuations, stopwords and get a cleaned list of words for each message, also called Vectorization
## this inclused tokenization, normalization, vectorization

Tokenization: Breaking the text into individual tokens (words or subwords).

Example: "I love AI" → ["I", "love", "AI"]
Normalization: Preprocessing the tokens to standardize them, such as lowercasing, removing stopwords, punctuation, and stemming.

Example: "I love AI!" → "love ai"
Vectorization: Converting tokens into numerical vectors using methods like Bag of Words, TF-IDF, or Word Embeddings.

Example (BoW for "love ai"): [0, 1, 1, 0, 0]
    
    


# 2. Then we create a bag of words for each message that is word with counts in the message using CountVectorizer

# 3. Then we create a tfidf for each bag of words that is each word in a message is given some weightage using
# TfidfTransformer

# 4. Finally naive bayes classification can be used to classify and predict.
```

## Train_test_split and proper way 

```python
from sklearn.model_selection import train_test_split
```

```python
messages_train, messages_test, label_train, label_test = train_test_split(messages["message"], messages["label"], test_size = 0.3, random_state = 42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ("bow_transformer", CountVectorizer(analyzer= word_processing)),
    ("tdidf_transformer", TfidfTransformer()),
    ("classfier", MultinomialNB())
])
```

```python
pipeline.fit(messages_train, label_train)
```

```python
predictions = pipeline.predict(messages_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(label_test, predictions))
```

```python
accuracy_score(label_test, predictions)
```

```python
confusion_matrix(label_test, predictions)
```
